```python
import pandas as pd
from dataclasses import dataclass
from typing import List

# Path file CSV
data_mahasiswa_path = r"C:\MATA KULIAH\AKADEMIK\SEMESTER 6\Data Science\week 6\Data_Mahasiswa.csv"
iris_dataset_path = r"C:\MATA KULIAH\AKADEMIK\SEMESTER 6\Data Science\week 6\IrisDataset.csv"

# Membaca dataset dengan delimiter yang benar (titik koma `;`)
df_mahasiswa = pd.read_csv(data_mahasiswa_path, encoding="utf-8", sep=";")
df_iris = pd.read_csv(iris_dataset_path, encoding="utf-8")

# Membersihkan nama kolom dari spasi ekstra
df_mahasiswa.columns = df_mahasiswa.columns.str.strip()

# Cek isi kolom sebelum konversi
print(df_mahasiswa[["Angkatan Mahasiswa"]].head(10))

# Konversi ke numeric, ubah non-numeric jadi NaN, lalu isi NaN dengan 2020
df_mahasiswa["Angkatan Mahasiswa"] = pd.to_numeric(df_mahasiswa["Angkatan Mahasiswa"], errors='coerce').fillna(2020).astype(int)

# Data Class untuk Data Mahasiswa
@dataclass
class Mahasiswa:
    email: str
    nama: str
    nim: str
    prodi: str
    angkatan: int

# Konversi Data Mahasiswa ke List of Objects
mahasiswa_list: List[Mahasiswa] = [
    Mahasiswa(
        row["Email Address"],
        row["Nama Lengkap Mahasiswa"],
        row["NIM"],
        row["Program Studi Mahasiswa"],
        row["Angkatan Mahasiswa"]  # Sudah dipastikan integer
    ) for _, row in df_mahasiswa.iterrows()
]

# Data Class untuk Dataset Iris
@dataclass
class Iris:
    sepal_length: float
    sepal_width: float
    petal_length: float
    petal_width: float
    species: str

# Konversi Data Iris ke List of Objects
iris_list: List[Iris] = [
    Iris(
        row["sepal_length"],
        row["sepal_width"],
        row["petal_length"],
        row["petal_width"],
        row["species"]
    ) for _, row in df_iris.iterrows()
]

# Cek hasil
print("Mahasiswa Pertama:", mahasiswa_list[0])
print("Data Iris Pertama:", iris_list[0])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path file CSV
data_mahasiswa_path = r"C:\MATA KULIAH\AKADEMIK\SEMESTER 6\Data Science\week 6\Data_Mahasiswa.csv"
iris_dataset_path = r"C:\MATA KULIAH\AKADEMIK\SEMESTER 6\Data Science\week 6\IrisDataset.csv"

# Membaca dataset dengan delimiter yang benar (titik koma `;`)
df_mahasiswa = pd.read_csv(data_mahasiswa_path, encoding="utf-8", sep=";")
df_iris = pd.read_csv(iris_dataset_path, encoding="utf-8")

# Membersihkan nama kolom dari spasi ekstra
df_mahasiswa.columns = df_mahasiswa.columns.str.strip()

# Konversi kolom "Angkatan Mahasiswa" menjadi numerik
df_mahasiswa["Angkatan Mahasiswa"] = pd.to_numeric(df_mahasiswa["Angkatan Mahasiswa"], errors='coerce').fillna(2020).astype(int)

# Mengatur style visualisasi
sns.set(style="whitegrid")

#  1 **Histogram Angkatan Mahasiswa**
plt.figure(figsize=(8, 5))
sns.countplot(x="Angkatan Mahasiswa", data=df_mahasiswa, palette="coolwarm")
plt.title("Distribusi Angkatan Mahasiswa")
plt.xlabel("Angkatan")
plt.ylabel("Jumlah Mahasiswa")
plt.xticks(rotation=45)
plt.show()

#  2 **Pie Chart Penggunaan Media Sosial**
media_laptop = df_mahasiswa["Apakah Anda biasanya menggunakan media sosial melalui laptop?"].value_counts()
media_ponsel = df_mahasiswa["Apakah Anda biasanya menggunakan media sosial melalui ponsel?"].value_counts()

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].pie(media_laptop, labels=media_laptop.index, autopct="%1.1f%%", colors=["skyblue", "lightcoral"])
ax[0].set_title("Penggunaan Media Sosial via Laptop")

ax[1].pie(media_ponsel, labels=media_ponsel.index, autopct="%1.1f%%", colors=["lightgreen", "orange"])
ax[1].set_title("Penggunaan Media Sosial via Ponsel")

plt.show()

#  3 **Scatter Plot Sepal vs Petal dari Dataset Iris**
plt.figure(figsize=(8, 6))
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=df_iris, palette="viridis")
plt.title("Scatter Plot Sepal vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()

#  4 **Histogram Sepal & Petal**
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df_iris["sepal_length"], bins=20, kde=True, color="blue", ax=ax[0])
ax[0].set_title("Distribusi Sepal Length")

sns.histplot(df_iris["petal_length"], bins=20, kde=True, color="red", ax=ax[1])
ax[1].set_title("Distribusi Petal Length")

plt.show()
```
