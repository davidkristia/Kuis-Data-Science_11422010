```python
from dataclasses import dataclass

@dataclass
class DataPoint:
    feature1: float
    feature2: float
    label: int

# Contoh penggunaan
data = DataPoint(5.3, 2.1, 1)
print(data)
```

```python
from dataclasses import dataclass
from typing import List

@dataclass
class Dataset:
    features: List[List[float]]
    labels: List[int]

# Contoh dataset
data = Dataset(
    features=[[5.3, 2.1], [3.2, 1.7], [4.1, 2.3]],
    labels=[0, 1, 2]
)
print(data)
```

```python
import torch
from torch.utils.data import Dataset

class LeafDataset(Dataset):
    def __init__(self, features, labels):
        self.features = torch.tensor(features, dtype=torch.float32)
        self.labels = torch.tensor(labels, dtype=torch.long)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]

# Contoh penggunaan
data_features = [[5.3, 2.1], [3.2, 1.7], [4.1, 2.3]]
data_labels = [0, 1, 2]
dataset = LeafDataset(data_features, data_labels)

# Akses satu data
print(dataset[0])
```
